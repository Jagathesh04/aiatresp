# Generating SDO/AIA Temperature Responses with aiatresp

This notebook demonstrates how to compute SDO/AIA temperature response functions using `aiatresp` without needing SolarSoft or IDL.

```python
from aiatresp import aia_response
import matplotlib.pyplot as plt
import numpy as np
```

## 1. Compute Temperature Response for an Observation Epoch

```python
# Compute response for 2017-01-01
resp = aia_response(obstime="2017-01-01T00:00:00", calibration_version=10)

print("Channels:", resp.channels)
print("Temperature grid size:", len(resp.logt))
print("Response matrix shape:", resp.response.shape)
```

## 2. Plotting Channel Responses

```python
fig, ax = plt.subplots(figsize=(9, 5))
for i, ch in enumerate(resp.channels):
    ax.plot(resp.logt, resp.response[i, :], label=f"{ch} Å")

ax.set_yscale("log")
ax.set_xlabel(r"$\log_{10}(T \ [\mathrm{K}])$")
ax.set_ylabel("Response [DN cm^5 s^-1 pixel^-1]")
ax.set_title("SDO/AIA Temperature Response (2017-01-01)")
ax.grid(True, which="both", linestyle=":", alpha=0.5)
ax.legend()
plt.show()
```
