# Using aiatresp with Differential Emission Measure (DEM) Analysis

This notebook shows how `aiatresp` response matrices interface directly with DEM inversion workflows.

```python
from aiatresp import aia_response
import numpy as np
import matplotlib.pyplot as plt
```

## 1. Load Response Matrix for DEM Kernel

```python
resp = aia_response(obstime="2017-01-01T00:00:00")
# Kernel matrix K where K[channel, temperature] * DEM(T) * dT = Intensity
kernel = resp.response
temps = 10**resp.logt

print("DEM Kernel Matrix Shape:", kernel.shape)
print("Temperature Grid Bounds:", temps[0], "K to", temps[-1], "K")
```
