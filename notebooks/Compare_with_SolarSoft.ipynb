# Comparing aiatresp with SolarSoft IDL

This notebook demonstrates scientific verification of `aiatresp` outputs against baseline SolarSoft IDL `aia_get_response` computations.

```python
from aiatresp import aia_response
import numpy as np
import matplotlib.pyplot as plt
```

## 1. Compute Python Response vs Single-Precision Baseline

```python
py_resp = aia_response(obstime="2017-01-01T00:00:00", calibration_version=10)

# IDL internal single-precision truncation model
idl_sim = py_resp.response.astype(np.float32).astype(np.float64)
abs_diff = np.abs(py_resp.response - idl_sim)

print(f"Max Absolute Error: {np.max(abs_diff):.6e}")
print(f"RMSE:               {np.sqrt(np.mean(abs_diff**2)):.6e}")
```
